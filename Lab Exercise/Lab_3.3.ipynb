```python
df_car.info()
```

```python
df_car['num-of-doors'].value_counts()
```

two types of value: two and four doors

```python
door_mapper = {"two": 2,
              "four": 4}
```

```python
df_car['doors'] = df_car["num-of-doors"].replace(door_mapper)
```

```python
df_car.head()
```

succesfully create new column for number of doors into numerical value

```python
df_car['num-of-cylinders'].value_counts()
```

eight types of value

```python
cylinder_mapper = {"two":2,
                  "three":3,
                  "four":4,
                  "five":5,
                  "six":6,
                  "eight":8,
                  "twelve":12}
```

```python
df_car['cylinders'] = df_car['num-of-cylinders'].replace(cylinder_mapper)
```

```python
df_car.head()
```

succesfully create new column for number of cylinders into numerical value

```python
df_car['drive-wheels'].value_counts()
```

three types of non-ordinal values

```python
df_car = pd.get_dummies(df_car,columns=['drive-wheels'])
```

```python
df_car.head()
```

succesfully create <b>three</b> new columns for drive wheels type with get_dummies function

```python
df_car['aspiration'].value_counts()
```

```python
df_car = pd.get_dummies(df_car,columns=['aspiration'], drop_first=True)
```
